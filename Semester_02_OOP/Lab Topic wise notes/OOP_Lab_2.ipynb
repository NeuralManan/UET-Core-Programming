## ðŸ“Œ ***Encapsulation***

The concept of hiding the internal state of an object and controlling access to it through methods (getters/setters). It prevents external code from directly manipulating internal variables in ways that could break the objectâ€™s state.

This is typically done by:

* Prefixing attributes with _ or __ (weak or strong â€œprivacyâ€).

* Using getter and setter methods or @property decorators to control access and enforce rules.

```python
class Animal:
    """
    A class to represent an animal with encapsulated attributes:
    - name
    - age
    - vaccination status

    Attributes are made 'private' with a leading double underscore and accessed via properties.
    """

    def __init__(self, name: str = "Camel", age: int = 12, vaccinated: bool = None):
        """
        Initializes an Animal instance.

        Args:
            name (str): Name of the animal. Default is "Camel".
            age (int): Age of the animal in years. Default is 12.
            vaccinated (bool | None): Vaccination status. Default is None (unknown).
        """
        self.__name: str = name  # Private attribute for name
        self.__age: int = age    # Private attribute for age
        self.__vaccinated: bool | None = vaccinated  # Private attribute for vaccination
        self._vaccination_status: str = "Vaccination status is not known yet"  # Internal status

        # Initialize the vaccination status string based on vaccinated attribute
        if vaccinated is not None:
            self.__update_vaccination_status()

    # --- Getter and setter for 'name' ---
    @property
    def name(self) -> str:
        """Gets the name of the animal."""
        return self.__name

    @name.setter
    def name(self, new_name: str):
        """Sets the name of the animal."""
        self.__name = new_name

    # --- Getter and setter for 'age' ---
    @property
    def age(self) -> int:
        """Gets the age of the animal."""
        return self.__age

    @age.setter
    def age(self, new_age: int):
        """Sets the age of the animal."""
        if new_age < 0:
            raise ValueError("Age cannot be negative.")
        self.__age = new_age

    # --- Getter and setter for 'vaccinated' ---
    @property
    def vaccinated(self) -> bool | None:
        """Gets the vaccination status of the animal."""
        return self.__vaccinated

    @vaccinated.setter
    def vaccinated(self, status: bool):
        """Sets the vaccination status and updates the vaccination string."""
        self.__vaccinated = status
        self.__update_vaccination_status()

    # --- Private method to update vaccination status string ---
    def __update_vaccination_status(self):
        """Updates the internal vaccination status string based on _vaccinated."""
        if self.__vaccinated is True:
            self._vaccination_status = "is vaccinated."
        elif self.__vaccinated is False:
            self._vaccination_status = "is not vaccinated."
        else:
            self._vaccination_status = "Vaccination status is not known yet."

    # --- String representations ---
    def __str__(self) -> str:
        """Human-readable string of the animal."""
        return f"{self.__name} is {self.__age} years old and {self._vaccination_status}"

    def __repr__(self) -> str:
        """Developer-friendly representation."""
        return f"Animal(name={self.__name}, age={self.__age}, vaccination_status={self._vaccination_status})"

    # --- Destructor (optional) ---
    def __del__(self):
        """Destructor called when object is deleted."""
        pass
```

âœ… Key Encapsulation Features

* Private attributes: __name, __age, __vaccinated

* Protected attribute: _vaccination_status.

* Controlled access: @property getters and setters.

* Internal logic protected: __update_vaccination_status() ensures consistency.

* Validation example: age cannot be negative.

```python
# Create an instance of the Animal class using default values
animal_obj = Animal()
# At this point:
#   name = "Camel" (default)
#   age = 12 (default)
#   vaccinated = None (unknown)
#   _vaccination_status = "Vaccination status is not known yet"
print(animal_obj)
# Calls animal_obj.__str__()
# Output: Camel is 12 years old and Vaccination status is not known yet

# Update the age using the setter
animal_obj.age = 30
# The setter validates and updates __age internally

# Update the name using the setter
animal_obj.name = "Fluffy Cat"
# The setter updates __name internally

# Update vaccination status using the setter
animal_obj.vaccinated = True
# The setter updates __vaccinated internally
# Also automatically calls _update_vaccination_status()
# _vaccination_status becomes "is vaccinated."

print(animal_obj)
# Calls __str__ again
# Output: Fluffy Cat is 30 years old and is vaccinated.

# Change name and vaccination status again
animal_obj.name = "Black Cobra"
# __name is updated internally

animal_obj.vaccinated = False
# __vaccinated is updated internally
# __vaccination_status becomes "is not vaccinated."

print(animal_obj)
# Output: Black Cobra is 30 years old and is not vaccinated.

# --- Optional/debug lines ---
# repr(animal_obj)
# Returns a developer-friendly representation of the object using __repr__:
# Output: Animal(name='Black Cobra', age=30, vaccination_status='is not vaccinated.')

# dir(animal_obj)
# Returns a list of all attributes and methods of the object
# Useful for debugging or inspecting the object

# del animal_obj
# Deletes the object and calls the __del__ method
# After this, animal_obj is no longer accessible
```

# ðŸ’¡ `hasattr`

`hasattr(object, 'attr')` is a built-in function that checks if an object has an attribute or property named `'attr'`.

- **Returns `True`** if the attribute exists (including properties defined with `@property`).  
- **Returns `False`** if the attribute does not exist.

```python
print(hasattr(animal_obj, 'name'))
print(hasattr(animal_obj, '_name'))
print(hasattr(animal_obj, 'age'))
print(hasattr(animal_obj, '__age'))
print(hasattr(animal_obj, '_Animal__age'))
print(hasattr(animal_obj, 'fage'))
```

# ðŸ“Œ ***Inheritance***

**Inheritance** is an OOP concept where a **child class** (subclass) derives properties and behaviors from a **base class** (parent class), allowing code reuse. The child class can **override or extend** the base class methods and attributes. This establishes a **hierarchical relationship** between classes, promoting modular and maintainable code.


# **Single Inheritance**

```python
# ðŸ“Œ Cat class inherits from Animal
class Cat(Animal):  # Inheritance: Cat is a subclass of Animal
    def speak(self):
        print("meow")  # Specific behavior for Cat objects
```

```python
# Create Cat object (Inheritance: __init__ inherited from Animal)
cat_obj = Cat("baby-cat", 15)

# Update name (Encapsulation: using setter)
cat_obj.name = "fluffy-cat"

# Print human-readable string (Inheritance: __str__ from Animal)
print(cat_obj)  # Output: fluffy-cat is 15 years old and Vaccination status is not known yet

# Call Cat-specific method
cat_obj.speak()  # Output: meow

# Access age via getter (Encapsulation)
print(cat_obj.age)  # Output: 15
```

# **Hierarchical Inheritance**

```python
# ðŸ“Œ Dog class inherits from Animal
class Dog(Animal):
    """
    Dog class inherits attributes and methods from Animal and adds dog-specific features.
    """

    def __init__(self, name: str, age: int, breed: str):
        # ðŸ“Œ Call the parent class constructor (Inheritance)
        super().__init__(name, age)  # Initializes __name, __age, and __vaccinated
        self.breed: str = breed      # Additional attribute specific to Dog

    # ðŸ“Œ Dog-specific behavior (Polymorphism)
    def speak(self):
        print("Woof!")  # Different behavior than other Animal subclasses

    # ðŸ“Œ Compare age with another Dog object
    def age_diff(self, other):
        if not isinstance(other, Dog):
            print("Can only compare age with another Dog")
            return

        if self.age > other.age:
            diff = self.age - other.age
            print(f"{self.name} is {abs(diff)} years older than {other.name}")
        elif self.age < other.age:
            diff = other.age - self.age
            print(f"{other.name} is {abs(diff)} years older than {self.name}")
        else:
            print(f"{self.name} and {other.name} are the same age")

    # ðŸ“Œ Human-readable string representation
    def __str__(self):
        return f"{self.name} is {self.age} years old, breed: {self.breed}, {self._vaccination_status}"
```

```python

print("\n---- Dog tests ----")

# ðŸ“Œ Create Dog objects (Inheritance: __init__ from Animal)
d1 = Dog("Buddy", 5, "Golden Retriever")
d2 = Dog("Max", 3, "Beagle")

# ðŸ“Œ Set vaccination status (Encapsulation: setter)
d1.vaccinated = True
d2.vaccinated = False

# ðŸ“Œ Print dogs (Inheritance: __str__ from Dog / Animal)
print(d1)  # Output: Buddy is 5 years old, breed: Golden Retriever, is vaccinated.
print(d2)  # Output: Max is 3 years old, breed: Beagle, is not vaccinated.

# ðŸ“Œ Call Dog-specific method (Polymorphism)
d1.speak()  # Output: Woof!

# ðŸ“Œ Compare ages using age_diff()
d1.age_diff(d2)  # Output: Buddy is 2 years older than Max
```

# **Multilevel Inheritance**

```python
# --------------------- PersianCat Class ---------------------
# ðŸ“Œ PersianCat inherits from Cat, which itself inherits from Animal. This is multi-level inheritance.
class PersianCat(Cat):
    """
    PersianCat class is a specialized Cat with an additional 'breed_type' attribute.
    """

    def __init__(self, name: str = "Camel", age: int = 12, breed_type: str = "Persian"):
        # ðŸ“Œ Call the parent class constructor (Cat -> Animal)
        super().__init__(name, age)  # Initializes __name, __age, __vaccinated
        self.breed_type: str = breed_type  # Specific to PersianCat

    # ðŸ“Œ Human-readable string representation
    def __str__(self) -> str:
        return (f"{self.name} is {self.age} years old, "
                f"breed: {self.breed_type}, {self._vaccination_status}")

    # ðŸ“Œ Polymorphic method: can override speak if desired
    def speak(self):
        print("meow (PersianCat)")  # Specific behavior for PersianCat
```

```python
print("\n---- PersianCat tests ----")

# Create PersianCat object
pcat = PersianCat("Fluffy", 3, "Himalyan Persian Cat")

# Access and update encapsulated attributes (via getters/setters)
pcat.name = "Fluffy Persian"
pcat.vaccinated = True

# Print PersianCat info (uses __str__ from PersianCat)
print(pcat)  # Output: Fluffy Persian is 3 years old, breed: Persian, is vaccinated.

# Call PersianCat-specific speak method (Polymorphism)
pcat.speak()  # Output: meow (PersianCat)
```

# **Multiple Inheritance**

```python
# --------------------- Bird Class ---------------------
class Bird:
    """
    A class to represent bird-specific behavior.
    """
    def __init__(self, wing_span: float = 1.0):
        self.wing_span = wing_span  # in meters

    def fly(self):
        print(f"This bird is flying with a wingspan of {self.wing_span} meters.")

    def speak(self):
        print("Chirp!")  # Default bird sound

# --------------------- Parrot Class (Multiple Inheritance) ---------------------
class Parrot(Animal, Bird):
    """
    Parrot inherits from both Animal and Bird.
    Demonstrates multiple inheritance.
    """
    def __init__(self, name: str = "Polly", age: int = 2, vaccinated: bool = None, wing_span: float = 0.5):
        Animal.__init__(self, name, age, vaccinated)  # Initialize Animal part
        Bird.__init__(self, wing_span)                # Initialize Bird part

    # Overriding speak to combine both behaviors
    def speak(self):
        print(f"{self.name} says Squawk! and can also chirp like a bird.")

    def move(self):
        # Parrot can walk (from Animal perspective) or fly (from Bird perspective)
        print(f"{self.name} is walking a little...")
        self.fly()  # call Bird's fly method

    def __str__(self):
        return f"{self.name} is {self.age} years old, wingspan: {self.wing_span} meters, {self._vaccination_status}"

# --------------------- Usage ---------------------
parrot1 = Parrot(name="Polly", age=3, vaccinated=True, wing_span=0.6)
print(parrot1)         # Shows combined info from Animal and Bird
parrot1.speak()         # Overridden speak method
parrot1.move()          # Move method demonstrates Animal + Bird behavior
```

# ðŸ“Œ ***Abstraction***

Abstraction using an `@abstractmethod` allows a class to **define a method interface without providing its implementation**, forcing all subclasses to implement that method.
This hides the **internal details** of how the method works, exposing only the **essential functionality** to the user.


```python
from abc import ABC, abstractmethod

class Animal(ABC):
    """
    Abstract base class representing an animal with encapsulated attributes:
    - name
    - age
    - vaccination status

    Attributes are private/protected and accessed via properties.
    """

    def __init__(self, name: str = "Camel", age: int = 12, vaccinated: bool = None):
        self.__name: str = name  # Private attribute
        self.__age: int = age    # Private attribute
        self.__vaccinated: bool | None = vaccinated  # Private attribute
        self._vaccination_status: str = "Vaccination status is not known yet"  # Protected for subclasses

        if vaccinated is not None:
            self.__update_vaccination_status()

    # --- Getter and setter for name ---
    @property
    def name(self) -> str:
        return self.__name

    @name.setter
    def name(self, new_name: str):
        self.__name = new_name

    # --- Getter and setter for age ---
    @property
    def age(self) -> int:
        return self.__age

    @age.setter
    def age(self, new_age: int):
        if new_age < 0:
            raise ValueError("Age cannot be negative.")
        self.__age = new_age

    # --- Getter and setter for vaccinated ---
    @property
    def vaccinated(self) -> bool | None:
        return self.__vaccinated

    @vaccinated.setter
    def vaccinated(self, status: bool):
        self.__vaccinated = status
        self.__update_vaccination_status()

    # --- Private method to update vaccination status string ---
    def __update_vaccination_status(self):
        if self.__vaccinated is True:
            self._vaccination_status = "is vaccinated."
        elif self.__vaccinated is False:
            self._vaccination_status = "is not vaccinated."
        else:
            self._vaccination_status = "Vaccination status is not known yet."

    # --- Abstract methods ---
    @abstractmethod
    def speak(self):
        """Abstract method for animal sound. Must be implemented by derived classes."""
        pass

    @abstractmethod
    def move(self):
        """Abstract method for movement behavior. Must be implemented by derived classes."""
        pass

    # --- String representations ---
    def __str__(self) -> str:
        return f"{self.__name} is {self.__age} years old and {self._vaccination_status}"

    def __repr__(self) -> str:
        return f"Animal(name={self.__name}, age={self.__age}, vaccination_status={self._vaccination_status})"

    # --- Destructor ---
    def __del__(self):
        pass
```

```python
# --------------------- Cat Subclass ---------------------
class Cat(Animal):
    """
    Cat class inherits from Animal.
    Implements abstract methods speak() and move().
    """
    def __init__(self, name: str = "Cat", age: int = 2, vaccinated: bool = None):
        super().__init__(name, age, vaccinated)
        self.color: str = "unknown"  # Cat-specific attribute

    # ðŸ“Œ Implementation of abstract method
    def speak(self):
        print("meow")

    # ðŸ“Œ Implementation of abstract method
    def move(self):
        print(f"{self.name} is walking gracefully.")

    def __str__(self):
        return f"{self.name} is {self.age} years old, color: {self.color}, {self._vaccination_status}"


# --------------------- Dog Subclass ---------------------
class Dog(Animal):
    """
    Dog class inherits from Animal.
    Implements abstract methods speak() and move().
    """
    def __init__(self, name: str = "Dog", age: int = 3, vaccinated: bool = None, breed: str = "Unknown"):
        super().__init__(name, age, vaccinated)
        self.breed: str = breed  # Dog-specific attribute

    # ðŸ“Œ Implementation of abstract method
    def speak(self):
        print("Woof!")

    # ðŸ“Œ Implementation of abstract method
    def move(self):
        print(f"{self.name} is running energetically.")

    def __str__(self):
        return f"{self.name} is {self.age} years old, breed: {self.breed}, {self._vaccination_status}"


# --------------------- Usage Example ---------------------
print("\n---- Animal Subclasses Tests ----")

# Create Cat object
cat1 = Cat("Whiskers", 3, vaccinated=True)
cat1.color = "White"
print(cat1)       # Whiskers is 3 years old, color: White, is vaccinated.
cat1.speak()      # meow
cat1.move()       # Whiskers is walking gracefully.

# Create Dog object
dog1 = Dog("Buddy", 5, vaccinated=False, breed="Golden Retriever")
print(dog1)       # Buddy is 5 years old, breed: Golden Retriever, is not vaccinated.
dog1.speak()      # Woof!
dog1.move()       # Buddy is running energetically.
```

```python
# --------------------- PersianCat Class ---------------------
# ðŸ“Œ PersianCat inherits from Cat, which itself inherits from Animal
class PersianCat(Cat):
    """
    PersianCat class is a specialized Cat with an additional 'breed_type' attribute.
    """

    def __init__(self, name: str = "Camel", age: int = 12, breed_type: str = "Persian"):
        # ðŸ“Œ Call the parent class constructor (Cat -> Animal)
        super().__init__(name, age)  # Initializes __name, __age, __vaccinated
        self.breed_type: str = breed_type  # Specific to PersianCat

    # ðŸ“Œ Human-readable string representation
    def __str__(self) -> str:
        return (f"{self.name} is {self.age} years old, "
                f"breed: {self.breed_type}, {self._vaccination_status}")

    # ðŸ“Œ Polymorphic method: can override speak if desired
    def speak(self):
        print("meow (PersianCat)")  # Specific behavior for PersianCat
```

# ðŸ“Œ ***Polymorphism***

`Run-time polymorphism` via method overriding occurs when a **subclass provides its own implementation of a method defined in its parent class**, and the **subclass version is called at runtime**.
This allows **different behaviors** for the same method name depending on the object type.


```python
animals = [Cat("Whiskers", 3, vaccinated=True), PersianCat("WhitePersian")]

for animal in animals:
    animal.speak()  # Works because all objects have speak(), no inheritance needed
```

`Dynamic Polymorphism (Duck Typing)`
occurs when Python calls methods on objects without caring about their type, as long as the object has the required method/behavior.

```python
class Parrot:
    def speak(self):
        print("Squawk!")

    def move(self):
        print("Parrot is flying")

animals = [Dog(), Cat(), Parrot()]  # Parrot is NOT a subclass of Animal

for animal in animals:
    animal.speak()  # Works because all objects have speak(), no inheritance needed
    animal.move()   # Works because all objects have move()
```
